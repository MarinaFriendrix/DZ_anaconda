## Задание 1
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, 
а возвращает общее количество фруктов на складе, при помощи использования kwargs

```python
def sum (**fruits):
    sum = 0
    for key, value in fruits.items():
        sum += value
    return sum

fruits = {
    'яблоки': 5,
    'апельсины': 10,
    'бананы': 2
}

print(sum(**fruits))
```

## Задание 2
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты -->
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions.


```python
costs = [100, 125, -90, 345, 655, -1, 0, 200]
new_costs = [i for i in costs if i>0]
print (new_costs)
```

```python
def sum1 (*args):
    s = 0
    for i in args:
        s +=i
    return s
print (sum1(*new_costs))
```
